```python
from torchvision import transforms
from PIL import Image
import mlflow
import json

model = mlflow.pytorch.load_model("models:/bb_classifier/2")

with open("./label_map_category.json", 'r') as f:
    label_map = json.load(f)

label_map_swap = {v: k for k, v in label_map.items()}
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
file_path = "./onion2.webp"
file_path = file_path.replace("\\","/")
```

```python
image = Image.open(file_path).convert("RGB")
image = transform(image)
images = image.unsqueeze(0).to("cuda")
outputs = model(images)
preds = outputs.argmax(dim=1).cpu().numpy()
label_map_swap[int(preds)]
```
